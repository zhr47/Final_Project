```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow
```

```python
# modelling
from sklearn.preprocessing import StandardScaler, MinMaxScaler 
from sklearn.model_selection import train_test_split, cross_val_score 
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn import metrics
from sklearn.model_selection import train_test_split

from sklearn.metrics import explained_variance_score, mean_squared_error, r2_score
```

```python
df = pd.read_csv("data/simple_data_for_ML.csv")
```

```python
df.price = df.price.str[1:-4]
df.price = df.price.str.replace(",", "")
df.price = df.price.astype('int64')
df = df.loc[df.price < 1000]
```

```python
X = df[['id','bedrooms','beds','accommodates',
        'latitude', 'longitude',
        'number_of_reviews',
        'review_scores_rating',
        'host_identity_verified',
        'host_is_superhost',
        'reviews_per_month',
        'host_listings_count',
        'amenities_count',
        'instant_bookable',
        'reviews_per_month']]
y = df['price'].astype(float).values.reshape(-1,1)
print (X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 42)
```

```python
X_train = X_train.drop("id", axis=1)
X_test_id = X_test.copy()
X_test = X_test.drop("id", axis=1)
```

```python
X_scaler = StandardScaler().fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
#y_train_categorical = to_categorical(y_train)
#y_test_categorical = to_categorical(y_test)
y_test
```

```python
X_train.shape
```

```python
# first, create a normal neural network with 2 inputs, 6 hidden nodes, and 2 outputs
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

num_inputs = X_train.shape[1]

model = Sequential()
model.add(Dense(units= 3 * num_inputs, activation='relu', input_dim=num_inputs))
model.add(Dense(units=1, activation='linear'))
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='mean_squared_error')
```

```python
# Fit the model to the training data
model.fit(
    X_train_scaled,
    y_train,
    epochs=100,
    shuffle=True,
    verbose=2)
```

```python
from sklearn.metrics import r2_score
```

```python
y_predicted = model.predict(X_test_scaled)
r2 = r2_score(y_test, y_predicted)
```

```python
r2
```

```python
deep_model = Sequential()
deep_model.add(Dense(units= 3 * num_inputs, activation='relu', input_dim=num_inputs))
deep_model.add(Dense(units= 3 * num_inputs, activation='relu'))
deep_model.add(Dense(units=1, activation='linear'))
```

```python
deep_model.summary()
```

```python
deep_model.compile(optimizer='adam',
                   loss='mean_squared_error')

deep_model.fit(
    X_train_scaled,
    y_train,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
y_predicted_deep = deep_model.predict(X_test_scaled)
r2 = r2_score(y_test, y_predicted_deep)
r2
```
